```pyspark
lines = sc.textFile("s3://spark-dataset-sandbox/datasets/words/starwars.txt")
```

```pyspark
lines.count()
```

```pyspark
lines.take(3)
```

```pyspark
words = lines.flatMap(lambda lines: lines.split()) \
             .map(lambda word: (word,1)) \
             .cache() \
             .reduceByKey(lambda x,y: (x+y)) \
             .collect()
```

```pyspark
for word in words:
    print(word)
```
