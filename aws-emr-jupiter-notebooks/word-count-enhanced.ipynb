# The problem 
is about, finding word counts in a book. 


#### steps
* load book data
* flatmap  and split words 
* eliminate white spaces, lower/uppercase issues w/regex

  

```pyspark
import re
```

The only difference from /spark-sandbox/word-count-enhanced.py file is below file URL. 
- lowercase
- only words
- no whitespace

```pyspark
lines = sc.textFile("s3://spark-dataset-sandbox/datasets/words/book.txt")
```

```pyspark
lines.count()
```

```pyspark
lines.take(10)
```

```pyspark
words = lines.flatMap(lambda lines: re.compile(r'\W+', re.UNICODE).split(lines.lower())) \
             .map(lambda word: (word,1)) \
             .cache() \
             .reduceByKey(lambda x,y: (x+y)) \
             .cache() \
             .map(lambda x: (x[1],x[0])) \
             .sortByKey() \
             .collect()
```

```pyspark
for word in words:
    print(word)
```
