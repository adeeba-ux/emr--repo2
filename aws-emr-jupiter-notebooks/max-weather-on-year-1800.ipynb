# The problem 
is about, finding maximum Celcius in year 1800. 


#### steps
* load weather data
* map to station-id, entryType, celcius 
* correct celcius data (-180 degrees is not appropriate)
* filter "TMAX" records

  

```pyspark
sc.version
```

### Weather Data (year 1800)
stationid,date,entryType,celsius,,,,

```pyspark
lines = sc.textFile("s3://spark-dataset-sandbox/datasets/weather/year-1800.txt")
```

```pyspark
lines.count()
```

```pyspark
lines.take(3)
```

```pyspark
weathers = lines.map(lambda lines: lines.split(",")) \
                .cache() \
                .filter(lambda x: "TMAX" in x) \
                .map(lambda x: (x[0], float(x[3])/10)) \
                .cache()
```

```pyspark
weathers.count()
```

```pyspark
weathers.take(2)
```

```pyspark
weathers = weathers.reduceByKey(lambda x,y: max(x,y))
```
